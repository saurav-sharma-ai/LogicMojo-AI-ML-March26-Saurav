```python
import numpy as np
import pandas as pd
import sklearn1
```

```python
#Task-1
iris = sklearn.datasets.load_iris()
features = iris.data
print(iris.feature_names)
print(features)
```

```python
#Task-2
features_mean = np.mean(features, axis = 0)
features_median = np.median(features, axis = 0)
features_std = np.std(features, axis = 0)
features_variance = np.var(features, axis = 0)
print("Mean :" ,features_mean)
print("Median :" ,features_median)
print("Std :" ,features_std)
print("Variance :" ,features_variance)
```

```python
#Task-3
highest_variability = np.max(features_std)
index_of_variability = np.argmax(features_std)
print("Highest variablity in features is of the one which has the highest standard deviation, which is :")
print(f"Feature name: {iris.feature_names[index_of_variability]}  Feature deviation: {highest_variability}")
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
#Task-1,2,3
np.random.seed(10)
a = np.random.randn(1000)
std_a = np.std(a)
mean_a = np.mean(a)
lower = mean_a - 2*std_a
upper = mean_a + 2*std_a
outliers = a[(a < lower) | (a > upper)]
print(outliers) 
print(len(outliers))
```
