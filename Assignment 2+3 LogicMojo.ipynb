## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
def segment_customer(customer):
    ratio = customer["spend"] / customer["income"]
    if ratio > 0.6:
        return "High Value"
    elif ratio >= 0.3 and ratio <= 0.6:
        return "Medium Value"
    elif ratio < 0.3:
        return "Low Value"

for i in customers:
    i["segment"]= segment_customer(i)
print(customers)
            
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

def clean_text(text):
    text = text.replace("@@@"," ").replace("###"," ").replace("!!!", "!").replace("???","?").strip().lower()
    words = text.split()
    cleaned_text = " ".join(words)
    return cleaned_text

cleaned = list(map(clean_text,texts))
print(cleaned)
````

## 3. Fraud Detection Rule Engine

You are given:

```python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]
```

Task:

1. Write a function is_fraud(transaction)
2. If amount is greater than 100000 or location is "Unknown" mark as fraud
3. Otherwise mark as safe
4. Use lambda with filter or map
5. Return only fraud transactions


````python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]

def is_fraud(transaction):
    if transaction["amount"] > 100000 or transaction["location"] == "Unknown":
        return "Fraud"
    else:
        return "Safe"

for t in transactions:
    t["status"] = is_fraud(t)
frauds = list(filter(lambda x:x["status"]=="Fraud", transactions))
print(frauds)
````

## 4. Feature Engineering

You are given:

```python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]
```

Task:

1. Write a function feature_engineer(record)
2. Add a new key salary_per_exp
3. salary_per_exp is salary divided by experience
4. Handle division by zero using conditional logic
5. Return updated dataset

---

````python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]

def feature_engineer(record):
    try:
        record["salary_per_exp"] = record["salary"] / record["experience"]
    except ZeroDivisionError:
        record["salary_per_exp"] = None
    return record

new_data = list(map(feature_engineer,data))
new_data
            
````

## 5. Custom Reduce Function

You are given:

```python
accuracies = [0.8, 0.75, 0.9, 0.85]
```

Task:

1. Use reduce function to calculate average accuracy
2. Do not use sum function
3. Show intermediate logic using a custom function


````python
accuracies = [0.8, 0.75, 0.9, 0.85]

from functools import reduce

def total_acc(x,y):
    return x + y

total = reduce(total_acc,accuracies)
average = round(total/ len(accuracies),2)
print(average)
````

````python
#Alternate way:
accuracies = [0.8, 0.75, 0.9, 0.85]

def reduce(data):
    total = 0
    for i in data:
        total += i
    average = round(total / len(data),2)
    return average
reduce(accuracies)
````

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
data = [10, 20, 30, 25, 40, 50]

def detect_increasing_sequence(data):
    current_seq = [data[0]]
    longest_seq = [data[0]]
    
    for i in range(1, len(data)):
        if data[i] > data[i-1]:
            current_seq.append(data[i])
        else:
            if len(current_seq) > len(longest_seq):
                longest_seq = current_seq
            current_seq = [data[i]]
    return longest_seq
detect_increasing_sequence(data)
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

def recommend(user_purchases, target_user):
    
    target_items = set(user_purchases[target_user])
    
    recommended = set()
    
    for user, items in user_purchases.items():
        if user != target_user:
            items_set = set(items)
            
            if target_items & items_set:
                recommended.update(items_set)

    recommended -= target_items
    return list(recommended)
    
recommend(user_purchases,"user1")
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score >= 50 and score <= 80:
        return decision_tree(score+10)
    elif score < 50:
        return "Rejected"

decision_tree(55)
````
